#### Implement the following functions: mean_squared_error, mean_absolute_error, r2_score.
#### Each of them should have 2 vector arguments and return the appropriate score.

```python
def mean_squared_error(y_pred, yTest):
    result = [(yTest[i] - y_pred[i]) ** 2 for i in range(len(yTest))]
    MSE = sum(result) / len(result)
    return MSE
print(mean_squared_error(np.array(y_pred), np.array(yTest)))


def mean_absolute_error(y_pred, yTest):
    result = [abs(y_pred[i] - (yTest[i])) for i in range(len(yTest))]
    MAE = sum(result) / len(result)
    return MAE
print(mean_absolute_error(np.array(y_pred), np.array(yTest)))


def r2_score(y_pred, yTest):
    yTest_mean = sum(yTest) / len(yTest)
    RSS = sum([(yTest[i] - y_pred[i]) ** 2 for i in range(len(yTest))])
    TSS = sum([(yTest[i] - yTest_mean) ** 2 for i in range(len(yTest))])
    R_2 = 1 - (RSS / TSS)
    return R_2
print(r2_score(np.array(y_pred), np.array(yTest)))
```
